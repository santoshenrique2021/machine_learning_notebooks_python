# Notebook destinado a divisão dos dados em treino e teste

## Objetivo

* Dividir os dados em treino e teste para um problema de classificação.
* Etapa anterior ao EDA e a modelagem.

```python
#Imports
##Bibliotecas
import pandas as pd
import numpy as np
##Eliminar os warnings
import warnings
warnings.filterwarnings("ignore")
##Ver todas as colunas do data frame
pd.set_option('display.max_columns', None)
##Ver todas as linhas do data frame
pd.set_option('display.max_rows', None)
#Divisão da amostra
from sklearn.model_selection import train_test_split
```

# Visão geral dos dados

```python
df_final = pd.read_csv("NOME_DO_ARQUIVO.csv", sep = 'XX')
#Visualização
df_final.head(4)
```

# 1 - Divisão dos dados

```python
x_m6 = df_final.drop('ALVO', axis=1) #conjunto de variáveis explicativas
y_m6 = df_final['ALVO'] #ALVO
```

```python
#A base de treino vai ter 70% dos dados
#A base de teste vai ter 30% dos dados
m6treino_x, m6teste_x, m6treino_y, m6teste_y = train_test_split(x_m6, y_m6, random_state = 1935, test_size = 0.3, stratify = y_m6)
```

## 1.1 - Base de treino

```python
#Base de treino completa
df_treino = pd.concat([m6treino_x, m6treino_y], axis=1)
#Visualização
df_treino.head(4)
```

```python
#Volumetria
df_treino.shape[0] #117.306
```

```python
#Frequência do ALVO
round(100 * df_treino['ALVO'].value_counts(True),2) #4.76
```

### Salvar em CSV

```python
#Salvar em CSV
df_treino.to_csv('df_treino_m6.csv', header=True, index=False, sep=';')
```

## 1.2 - Base de teste

```python
#Base de treino completa
df_teste = pd.concat([m6teste_x, m6teste_y], axis=1)
df_teste.head(4)
```

```python
#Volumetria
df_teste.shape[0] #50274
```

```python
#Frequência do ALVO
round(100 * df_teste['ALVO'].value_counts(True),2) #4.77
```

### Salvar em CSV

```python
#Salvar em CSV
df_teste.to_csv('df_teste_m6.csv', header=True, index=False, sep=';')
```
